# **Accuracy 0.91 - 1 feauture**

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Embedding, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


X_text = result['description'].tolist()
y = result['numero'].tolist()

# Convertir clases a números
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Tokenizar los datos
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_text)
X = tokenizer.texts_to_sequences(X_text)
X = np.array(X)


# Definir una longitud máxima para las secuencias
max_sequence_length = 100

# Paddear las secuencias para que tengan la misma longitud
X = pad_sequences(X, maxlen=max_sequence_length)


# Construir el modelo LSTM
model = Sequential([
    Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=32, input_length=X.shape[1]),
    Conv1D(filters=64, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=128, kernel_size=3, activation='relu'),
    Dropout(0.4),
    GlobalMaxPooling1D(),
    Dense(units=len(set(y)), activation='softmax')
])


model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Entrenamiento del modelo
history = model.fit(X, np.array(y), epochs=15, batch_size=1, validation_split=0.2)

# Evaluar el modelo
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'Accuracy en datos de prueba: {accuracy}')


# Obtener métricas de entrenamiento
train_loss = history.history['loss']
train_accuracy = history.history['accuracy']

# Obtener métricas de validación
val_loss = history.history['val_loss']
val_accuracy = history.history['val_accuracy']

# Crear gráficos
plt.figure(figsize=(10, 4))

# Gráfico de pérdida (Loss)
plt.subplot(1, 2, 1)
plt.plot(train_loss, label='train')
plt.plot(val_loss, label='val')
plt.title('Loss')
plt.legend()

# Gráfico de precisión (Accuracy)
plt.subplot(1, 2, 2)
plt.plot(train_accuracy, label='train')
plt.plot(val_accuracy, label='val')
plt.title('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
